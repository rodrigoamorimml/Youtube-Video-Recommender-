## Get the Data

*To get the data, we'll use a library called **youtube_dlc**, this library get for us the videos and some information about them.*

```python
# import libraries

import pandas as pd
import numpy as np
import json, re, time
import youtube_dlc
```

```python
# We need to initialize the object
ydl = youtube_dlc.YoutubeDL({"ignoreerrors" : True}) # this argument set to true ignore possibles video that can't be downloaded
```

```python
# we'll be using some keywords for the video recommender
queries = [ "machine+learning", "kaggle", "deep+learning"]

resultados = []
for query in queries:
    r = ydl.extract_info(f'ytsearchdate300:{query}', download = False) # The first argument implies in get the information 
                                                                       # 300 videos for each query
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
        resultados += r['entries']
            
```

```python
len(resultados)
```

```python
resultados = [e for e in resultados if e is not None] # Retira os exemplos que deram algum erro para baixar
```

```python
print(f'Quantidade de Exemplos: {len(resultados)}')
```

```python
df = pd.DataFrame(resultados)
pd.set_option("display.max_columns", 66)
df.head()
```
